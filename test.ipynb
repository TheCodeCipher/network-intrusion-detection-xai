```python
import pandas as pd
import numpy as np

# Load just 5000 rows — fast to process
df = pd.read_csv('data/processed/cleaned_data.csv', nrows=5000)

# Check what attack types are in these rows
print(df['Label'].value_counts())
```

```python
# Load full dataset but only keep interesting rows
df_full = pd.read_csv('data/processed/cleaned_data.csv')

# Get 500 BENIGN rows
benign = df_full[df_full['Label'] == 'BENIGN'].sample(500, random_state=42)

# Get 200 DDoS rows
ddos = df_full[df_full['Label'] == 'DDoS'].sample(200, random_state=42)

# Get 100 rows each of other attacks
others = df_full[~df_full['Label'].isin(['BENIGN','DDoS'])].sample(300, random_state=42)

# Combine — mix them up so it looks like real traffic
demo_csv = pd.concat([benign, ddos, others]).sample(frac=1, random_state=42)
demo_csv.reset_index(drop=True, inplace=True)

print(f"Total rows: {len(demo_csv)}")
print(demo_csv['Label'].value_counts())

# Save it
demo_csv.to_csv('demo/demo_upload.csv', index=False)
print("Saved: demo/demo_upload.csv")
```
